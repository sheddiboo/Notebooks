**Importing Dependecies**

```python
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

**Data Collection**

```python
# Loading digits dataset from sklearn library
digit = load_digits()
```

```python
# feature name
digit.feature_names
```

```python
#Target name
digit.target_names
```

```python
# Storing the feautures in pandas datafram
df = pd.DataFrame(digit.data,columns=digit.feature_names)
df.head()
```

```python
# adding target to the dataframe 
df['target'] = digit.target
df.head()
```

```python
# checking the number of rows and columns
df.shape
```

**Seperating Features and Target**

```python
X = df.drop('target', axis=1)
Y = df['target']
```

```python
print(X)
```

```python
print(Y)
```

**Seperating thr Training and Test data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, random_state=13, stratify=Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model training: Support Vector Machines**

```python
model = SVC()
```

```python
model.fit(X_train, Y_train)
```

**Model Evaluation**

```python
# prediction on test data 
X_test_prediction = model.predict(X_test)

# accuracy score of test data 
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print("Accuracy Score on Test data :", test_data_accuracy)
```
