## Extrovert vs. Introvert Personality Traits Dataset

This dataset provides behavioral and social indicators to help distinguish between **extroverted** and **introverted** personality traits. It captures various lifestyle and social preferences, offering insight into personal tendencies and social energy levels.

### ðŸ“„ Dataset Features

| Feature Name               | Description                                                                 |
|---------------------------|-----------------------------------------------------------------------------|
| `Time_spent_Alone`        | Number of hours spent alone daily (range: 0â€“11)                             |
| `Stage_fear`              | Presence of stage fright (`Yes` / `No`)                                     |
| `Social_event_attendance` | Frequency of attending social events (range: 0â€“10)                          |
| `Going_outside`           | Frequency of going outside for non-essential reasons (range: 0â€“7)           |
| `Drained_after_socializing` | Feeling drained after socializing (`Yes` / `No`)                           |
| `Friends_circle_size`     | Number of close friends (range: 0â€“15)                                       |
| `Post_frequency`          | Frequency of posting on social media platforms (range: 0â€“10)               |
| `Personality`             | Target variable: Personality type (`Extrovert` / `Introvert`)              |

#### Exploratory Data Analysis

```python
personality_data.describe()
```

 1. Distribution of Each Feature

```python
sns.histplot(personality_data['Time_spent_Alone'], kde=True)
```

```python
# Set up the 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1
sns.histplot(personality_data['Social_event_attendance'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution: Social Event Attendance')

# Plot 2
sns.histplot(personality_data['Going_outside'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution: Going Outside Frequency')

# Plot 3
sns.histplot(personality_data['Friends_circle_size'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution: Friends Circle Size')

# Plot 4
sns.histplot(personality_data['Post_frequency'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution: Post Frequency')

# Adjust layout 
plt.tight_layout()
plt.show()
```

2. Compare Averages by Personality Type

```python
sns.barplot(x='Personality', y='Time_spent_Alone', data=personality_data)
```

```python
# Set figure size
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1
sns.barplot(x='Personality', y='Social_event_attendance', data=personality_data, ax=axes[0, 0])
axes[0, 0].set_title('Social Event Attendance by Personality')

# Plot 2
sns.barplot(x='Personality', y='Going_outside', data=personality_data, ax=axes[0, 1])
axes[0, 1].set_title('Going Outside Frequency by Personality')

# Plot 3
sns.barplot(x='Personality', y='Friends_circle_size', data=personality_data, ax=axes[1, 0])
axes[1, 0].set_title('Friends Circle Size by Personality')

# Plot 4
sns.barplot(x='Personality', y='Post_frequency', data=personality_data, ax=axes[1, 1])
axes[1, 1].set_title('Post Frequency by Personality')

# Adjust layout
plt.tight_layout()
plt.show()
```

 3. Correlation Between Features

```python
# Select only numeric columns
numeric_data = personality_data.select_dtypes(include='number')

plt.figure(figsize=(10, 5))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Numeric Personality Features")
plt.show()
```
