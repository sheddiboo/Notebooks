```python
# Filling the missing values in nations column(of df3) mapped with nation in df2
# Ensure both player columns have the same name and consistent formatting
df3['Player'] = df3['Player'].str.strip()
df2['player'] = df2['player'].str.strip()

# Merge df3 with nation info from df2
nation_map = df2[['player', 'nation']].dropna().drop_duplicates().set_index('player')['nation']

# Fill missing Nation values in df3
df3['Nation'] = df3.apply(
    lambda row: nation_map[row['Player']] if pd.isna(row['Nation']) and row['Player'] in nation_map else row['Nation'],
    axis=1
)
print("Remaining missing Nation values:", df3['Nation'].isna().sum())
```

#### Exploratory Data Analysis

1. Team Performance Summary

```python
# Compare goals scored vs. goals conceded across all teams 
df5_sorted = df5.sort_values("goals", ascending=True)
df5_sorted.plot(x='team', y=['goals', 'conceded'], kind='barh', figsize=(12, 6))
plt.title("Team Goals vs. Goals Conceded")
plt.ylabel("Goals")
plt.xticks(rotation=45)
plt.show()
```


2. Possession vs. Points

```python
# Examine whether more possession leads to better league performance 
merged = df5.merge(df6[['team', 'possession']], on='team')
sns.scatterplot(data=merged, x='possession', y='points')
plt.title("Team Possession vs. League Points")
plt.show()
```


3. Average Age vs. Performance

```python
# Does team age relate to league points or discipline 
age_stats = df8[['team', 'age']].drop_duplicates().merge(df5[['team', 'points']], on='team')
sns.barplot(x='team', y='age', data=age_stats)
plt.xticks(rotation=90)
plt.title("Team Average Age Distribution")
plt.show()
```


4. Referee Impact on Attendance

```python
# Compare average match attendance per referee
ref_attendance = df1.groupby('Referee')['Attendance'].mean().sort_values(ascending=True).head(10)
ref_attendance.plot(kind='barh', figsize=(10, 5), title='Avg Attendance by Referee')
plt.ylabel("Average Attendance")
plt.show()
```


5. Match Score Distribution

```python
# Explore the distribution of total goals in matches
df1['total_goals'] = df1['HomeScore'] + df1['AwayScore']
sns.histplot(df1['total_goals'], bins=20, kde=True)
plt.title("Distribution of Total Match Goals")
plt.xlabel("Goals per Match")
plt.show()
```


6. Salary vs. Team Performance

```python
# Relationship between team salary and league performance
salary_perf = df7.merge(df5[['team', 'points']], on='team')
sns.regplot(x='annual', y='points', data=salary_perf)
plt.title("Annual Team Salary vs. League Points")
plt.xlabel("Annual Salary (Million?)")
plt.ylabel("Points")
plt.show()
```


8. **Top Goal Scorers & Assists Providers**

```python
top_goals = df4[['name', 'goals']].sort_values('goals', ascending=False).head(10)
top_assists = df4[['name', 'assists']].sort_values('assists', ascending=False).head(10)
df4['total_contributions'] = df4['goals'] + df4['assists']
top_contrib = df4[['name', 'goals', 'assists', 'total_contributions']].sort_values('total_contributions', ascending=False).head(10)

# Plot setup
fig, axes = plt.subplots(3, 1, figsize=(10, 18)) 

# Top Goal Scorers
sns.barplot(ax=axes[0], data=top_goals, x='goals', y='name')
axes[0].set_title("Top 10 Goal Scorers")
axes[0].set_xlabel("Goals")
axes[0].set_ylabel("Player")
for i, value in enumerate(top_goals['goals']):
    axes[0].text(value + 0.3, i, str(int(value)), va='center', color='black')

# Top Assist Providers
sns.barplot(ax=axes[1], data=top_assists, x='assists', y='name')
axes[1].set_title("Top 10 Assist Providers")
axes[1].set_xlabel("Assists")
axes[1].set_ylabel("Player")
for i, value in enumerate(top_assists['assists']):
    axes[1].text(value + 0.3, i, str(int(value)), va='center', color='black')

# Top Goal Involvements
sns.barplot(ax=axes[2], data=top_contrib, x='total_contributions', y='name')
axes[2].set_title("Top 10 Goal Involvements (Goals + Assists)")
axes[2].set_xlabel("Total Contributions")
axes[2].set_ylabel("Player")
for i, value in enumerate(top_contrib['total_contributions']):
    axes[2].text(value + 0.3, i, str(int(value)), va='center', color='black')

plt.tight_layout()
plt.show()
```


9. Stadium Match Attendance

```python
# which stadiums draw the biggest crowds 
venue_attendance = df1.groupby('Venue')['Attendance'].mean().sort_values(ascending=True).head(10)
venue_attendance.plot(kind='barh', title='Average Attendance by Venue', figsize=(10, 5))
plt.ylabel("Avg Attendance")
plt.xticks(rotation=45)
plt.show()
```


10. Weekly Performance Trends (Line Plot)

```python
# Sum of goals scored each week
weekly = df1.groupby('week')[['HomeScore', 'AwayScore']].sum()
weekly['TotalGoals'] = weekly['HomeScore'] + weekly['AwayScore']

# Plot weekly goal trend
plt.figure(figsize=(10, 5))
sns.lineplot(data=weekly, x=weekly.index, y='TotalGoals', marker='o')
plt.title("Weekly Total Goals Scored Over The Season")
plt.xlabel("Week")
plt.ylabel("Goals")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Total Yellow and Red cards**

```python
# Calculate total yellow and red cards
total_yellow = df4['yellow'].sum()
total_red = df4['red'].sum()

# Values and labels
values = [total_yellow, total_red]
labels = ['Yellow Cards', 'Red Cards']
colors = ['gold', 'red']

# Plot
plt.figure(figsize=(6, 5))
bars = plt.bar(labels, values, color=colors)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 2, str(int(height)), 
             ha='center', va='bottom', fontsize=12)

plt.title("Total Yellow and Red Cards in the Season")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

**Yellow/Red Cards per Team**

```python
team_cards = df8.groupby('team')[['yellows', 'reds']].sum().reset_index()
team_cards = team_cards.sort_values('yellows', ascending=True)

team_cards.plot(x='team', kind='barh', stacked=True, figsize=(12, 6), color=['gold', 'red'])
plt.title("Yellow and Red Cards per Team")
plt.ylabel("Number of Cards")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Distribution of Red Cards**

```python
# Group and filter
team_reds = df8.groupby('team')['reds'].sum()
team_reds = team_reds[team_reds > 0]

# Calculate percentages
reds_pct = (team_reds / team_reds.sum()) * 100

# Sort for better visuals
reds_pct = reds_pct.sort_values()

# Plot horizontal bar chart
plt.figure(figsize=(10, 8))
bars = plt.barh(reds_pct.index, reds_pct.values, color=sns.color_palette('Reds', len(reds_pct)))

# Annotate bars with percentage
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2,
             f'{width:.1f}%', va='center')

plt.title("Percentage Distribution of Red Cards by Team")
plt.xlabel("Percentage (%)")
plt.tight_layout()
plt.show()
```
